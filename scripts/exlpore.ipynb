### Reading Pipe-Delimited .txt Files (e.g., Trade, DailyMarket, Account)

```python
import os
import pandas as pd
from IPython.display import display

DATA_DIR = r"C:\Users\Mega Store\Desktop\projects\Brokerage-Intelligence-Platform\data"
BATCH1_DIR = os.path.join(DATA_DIR, "Batch1")

txt_files = [
    "Trade.txt",
    "TradeHistory.txt",
    "HoldingHistory.txt",
    "CashTransaction.txt",
    "DailyMarket.txt",
    "WatchHistory.txt",
    "BatchDate.txt",
    "Date.txt",
    "Time.txt",
    "Industry.txt",
    "StatusType.txt",
    "TradeType.txt",
    "TaxRate.txt"
]

print(f"Starting to load {len(txt_files)} text files from Batch1...\n")

for file_name in txt_files:
    file_path = os.path.join(BATCH1_DIR, file_name)
    
    if os.path.exists(file_path):
        try:
            df = pd.read_csv(file_path, sep="|", header=None)
            
            print("=" * 80)
            print(f"📄 FILE: {file_name}")
            print(f"📊 SHAPE: {df.shape[0]} rows x {df.shape[1]} columns")
            print("=" * 80)
            
            display(df.head(5))
            print("\n" + "-" * 80 + "\n")
            
        except Exception as e:
            print(f"❌ Error loading {file_name}: {e}\n")
    else:
        print(f"⚠️ Warning: File '{file_name}' not found in Batch1 directory.\n")
```

### Reading Fixed-Width FINWIRE Files

```python
# Let's read one of the FINWIRE files as an example
finwire_path = os.path.join(BATCH1_DIR, "FINWIRE1967Q1")

companies = []

with open(finwire_path, 'r') as file:
    for line in file:
        # Check the record type in the first 15 characters (usually padding exists)
        record_type = line[15:18].strip()
        
        # If it is a Company (CMP) record, parse based on fixed character widths
        if record_type == "CMP":
            pts = line[0:15].strip() # Point in time
            company_name = line[18:78].strip()
            cik = line[78:88].strip()
            
            companies.append({
                'PTS': pts,
                'CompanyName': company_name,
                'CIK': cik
            })

df_companies = pd.DataFrame(companies)
print(f"Parsed {len(df_companies)} Company records from FINWIRE.")
display(df_companies.head())
```

### Parsing the Complex XML File (CustomerMgmt.xml)

```python
xml_path = os.path.join(BATCH1_DIR, "CustomerMgmt.xml")

# Parse the XML tree
tree = ET.parse(xml_path)
root = tree.getroot()

# Define a list to store parsed records
customer_actions = []

# Loop through each Action element in the XML
# The namespace is usually required for TPC-DI XML files
ns = {'tpcdi': 'http://www.tpc.org/tpc-di'}

for action in root.findall('.//tpcdi:Action', ns):
    action_type = action.attrib.get('ActionType')
    action_ts = action.attrib.get('ActionTS')
    
    # Extract customer details inside the action
    customer = action.find('tpcdi:Customer', ns)
    if customer is not None:
        cust_id = customer.attrib.get('C_ID')
        tax_id = customer.attrib.get('C_TAX_ID')
        
        # Get nested elements like Name, Address, Contact Info
        name_node = customer.find('tpcdi:Name', ns)
        first_name = name_node.find('tpcdi:C_F_NAME', ns).text if name_node is not None else None
        last_name = name_node.find('tpcdi:C_L_NAME', ns).text if name_node is not None else None
        
        # Append flat dictionary to our list
        customer_actions.append({
            'ActionType': action_type,
            'ActionTS': action_ts,
            'CustomerID': cust_id,
            'TaxID': tax_id,
            'FirstName': first_name,
            'LastName': last_name
        })

# Convert the parsed list of dicts to a Pandas DataFrame
df_customers = pd.DataFrame(customer_actions)

print(f"Parsed {len(df_customers)} customer action records from XML.")
display(df_customers.head(3))
```
